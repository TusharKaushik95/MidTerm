```python
class PriorityQueue:

    def __init__(self):

        self.queue = []

    def insert(self, item):

        self.queue.append(item)

    def min(self):

        return min(self.queue)

    def removeMin(self):

        self.queue.remove(self.min())

    def isEmpty(self):

        return len(self.queue) == 0

    def __len__(self):

        return len(self.queue)

    def __iter__(self):

        return self
    
    def isEmpty(self):
        return len(self.queue)==0
    
pq = PriorityQueue()

pq.insert(3)

pq.insert(1)

pq.insert(5)

pq.insert(2)

print('\nThe queue after inserting all the element are as follows:')

print(pq.queue)

print('\nMinimum element in the queue is '),

print(pq.min())

print('\nRemoving minimum element from the queue')

pq.removeMin()

print('\nAfter removing minimum element from the queue the number are as follows')

print(pq.queue)

print('\nLength of Queue is '),

print(pq.__len__())

print('\nIs the queue empty ')
print(pq.isEmpty())
```
